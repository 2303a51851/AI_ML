```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu=100
sigma=5
data = np.random.normal(mu,sigma,10000)
x=np.linspace(mu-3*sigma,mu+3*sigma,1000)
x=np.linspace(mu-3*sigma,mu+3*sigma,1000)
pdf=norm.pdf(x,mu,sigma)
plt.figure(figsize=(10,6))
plt.hist(data,bins=30,density=True,alpha=0.6,color='blue',edgecolor='black',linestyle='dashed',linewidth=1.5)
plt.plot(x,pdf,'r-',lw=2)
plt.title('Normal Distribution:$\mu$=100,$\sigma$=5')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x1=-1
x2=1
x=np.linspace(-4,4,1000)
y=stats.norm.pdf(x)


plt.plot(x,y,label='Normal Distribution')
plt.fill_between(x,y,where=(x>x1)&(x<x2),color='gray',alpha=0.5, label='p(x1<X<x2)')
plt.fill_between(x,y,where=(x<x2),color='blue',alpha=0.3, label='p(X<x2)')
plt.fill_between(x,y,where=(x>x1),color='green',alpha=0.3, label='p(x1<X)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution with shaded area')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_probabilities(IQ1,IQ2):
    mean=100
    std_dev=25

    P_ge_IQ1 = 1-stats.norm.cdf(IQ1,mean,std_dev)
    P_le_IQ1 = stats.norm.cdf(IQ1, mean, std_dev)
    P_between_IQ1_IQ2 = stats.norm.cdf(IQ2,mean,std_dev) - stats.norm.cdf(IQ1,mean,std_dev)

    print(f"P(X >= {IQ1}) = {P_ge_IQ1:.4f}")
    print(f"P(X <= {IQ1}) = {P_le_IQ1: .4f}")
    print(f"P({IQ1}<=x<={IQ2} = {P_between_IQ1_IQ2: .4f}")

    x=np.linspace(mean - 4*std_dev, mean+4*std_dev,1000)
    y=stats.norm.pdf(x,mean, std_dev)

    plt.plot(x,y,label='Normal Distribution')
    plt.fill_between(x,y,where=(x>=IQ1),color='red',alpha=0.5,label=f'P(X>={IQ1}')
    plt.fill_between(x,y,where=(x<=IQ1),color='blue',alpha=0.5,label=f'P(X<={IQ1}')
    plt.fill_between(x,y,where=(x>=IQ1)&(x<= IQ2),color='green',alpha=0.5,label=f'P({IQ1}<=x<={IQ2})')


    plt.xlabel('IQ')
    plt.ylabel('Probability Density')
    plt.title('Normal Distribution of IQ whith shaded Areas')
    plt.legend()
    plt.grid(True)
    plt.show()
IQ1=float(input("enter IQ1"))
IQ2=float(input("enter IQ2"))

calculate_probabilities(IQ1,IQ2)
```
