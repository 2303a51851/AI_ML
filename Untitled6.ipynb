```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
N=500
children_counts = np.random.choice([0, 1, 2, 3, 4], size=N, p=[0.1, 0.25, 0.35, 0.2, 0.1])
unique, counts = np.unique(children_counts, return_counts=True)
pmf=counts/N
cdf =np.cumsum(pmf)
plt .figure(figsize=(10,5))
sns.barplot(x=unique, y=pmf, color="red", edgecolor="black")
plt.xlabel("number of children per family")
plt.ylabel("probablity(pmf)")
plt.title("probabity mass function(pmf)of family sizes")
plt.xticks(unique)
plt.show()

plt .figure(figsize=(10,5))
sns.lineplot(x=unique, y=cdf,marker="o", color="red", linestyle="-")
plt.xlabel("number of children per family")
plt.ylabel("probablity(pmf)")
plt.title("probabity mass function(cdf)of family sizes")
plt.xticks(unique)
plt.ylim(0,1.1)
plt.show()





```
